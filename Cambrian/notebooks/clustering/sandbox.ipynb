```python
def json_to_dataframe_optimized(json_data):
    # Convert the nested JSON structure to a flat dictionary
    flat_data = {
        'article_id': [],
        'ah': [],
        'bh': [],
        'ar': [],
        'br': [],
        'cluster_id': []
    }
    
    for article_id, article_data in json_data.items():
        flat_data['article_id'].append(article_id)
        flat_data['ah'].append(article_data['interaction_data']['historic']['a'])
        flat_data['bh'].append(article_data['interaction_data']['historic']['b'])
        flat_data['ar'].append(article_data['interaction_data']['realtime']['ar'])
        flat_data['br'].append(article_data['interaction_data']['realtime']['br'])
        flat_data['cluster_id'].append(article_data['cluster_id'])
        
    # Create DataFrame in one go
    return pd.DataFrame(flat_data)
```

```python
def json_to_dataframe(json_data):
    # Initialize empty lists for storing data
    article_ids = []
    ah_values = []
    bh_values = []
    ar_values = []
    br_values = []
    cluster_ids = []
    
    # Loop through each article to gather data
    for article_id, article_data in json_data.items():
        article_ids.append(article_id)
        
        ah_values.append(article_data['interaction_data']['historic']['a'])
        bh_values.append(article_data['interaction_data']['historic']['b'])
        
        ar_values.append(article_data['interaction_data']['realtime']['ar'])
        br_values.append(article_data['interaction_data']['realtime']['br'])
        
        cluster_ids.append(article_data['cluster_id'])
    
    # Create DataFrame
    df = pd.DataFrame({
        'article_id': article_ids,
        'ah': ah_values,
        'bh': bh_values,
        'ar': ar_values,
        'br': br_values,
        'cluster_id': cluster_ids
    })
    
    df['a'] = df['ah'] + df['ar']
    df['b'] = df['bh'] + df['br']
    
    return df
```

```python
def _sample_probabilities_for_priors(article_prior_df):
    # Directly sample from Beta distribution using NumPy's vectorized operations
    sample_probabilities = np.random.beta(article_prior_df['a'], article_prior_df['b'])
    article_prior_df['sample_probability'] = sample_probabilities
    return article_prior_df
```

```python
def select_top_articles(df, cluster_doc_counts):
    # Step 1: Calculate the rank of each row within its cluster, when sorted by 'sampled_probability'
    df['rank'] = df.groupby('cluster_id')['sample_probability'].rank(method='first', ascending=False)
    
    # Step 2: Convert cluster_doc_counts to a DataFrame for easy merging
    cluster_limits_df = pd.DataFrame(list(cluster_doc_counts.items()), columns=['cluster_id', 'limit'])
    cluster_limits_df['cluster_id'] = cluster_limits_df['cluster_id'].astype(str)  # Make sure the types match
    
    # Merge the limit values into the main DataFrame
    df = pd.merge(df, cluster_limits_df, on='cluster_id', how='left')
    
    # Use boolean masking to select only the rows where the rank is <= the limit
    top_articles = df[df['rank'] <= df['limit']].drop(columns=['rank', 'limit']).reset_index(drop=True)
    
    return top_articles
```

```python
def select_top_articles(df, cluster_doc_counts):
    # Initialize an empty list to collect the DataFrames for each cluster
    dfs = []
    
    # Loop through each unique cluster_id in df
    for cluster_id, count in cluster_doc_counts:
        if count == 0:
            continue
        # Create a mask to select only rows with the current cluster_id
        mask = df['cluster_id'] == cluster_id
        cluster_df = df[mask]
        
        # Select the top 'n' rows for the current cluster_id, where 'n' is defined in cluster_doc_counts
        n = int(cluster_doc_counts.get(cluster_id, 0))
        top_n = cluster_df.nlargest(n, 'sample_probability')
        print(top_n)
        
        # Append the DataFrame for the current cluster to the list
        dfs.append(top_n)
    
    # Concatenate the list of DataFrames to create a single DataFrame
    top_articles = pd.concat(dfs, ignore_index=True)
    
    return top_articles
```

```python
def select_top_articles3(df, cluster_doc_counts):
    # Group the DataFrame by cluster_id and then apply the selection function for each group
    top_articles = df.groupby('cluster_id').apply(lambda group: group.nlargest(int(cluster_doc_counts.get(group.name, 0)), 'sample_probability'))
    # Remove the extra group-by index and reset the DataFrame index
    top_articles.reset_index(drop=True, inplace=True)
    top_articles.sort_values(by='sample_probability', ascending=False, inplace=True)
    return top_articles
```
